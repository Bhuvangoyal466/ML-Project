```python
import cv2 # open cv library
import pandas as pd # used for data
from ultralytics import YOLO 
import cvzone
```

```python
from ultralytics import YOLO

model = YOLO("best.pt")  # Load trained model
```

```python
model = YOLO('best.pt') # load the model
print(model.names)  # Prints the class labels the model detects
```

```python
cv2.namedWindow('RGB') # create a window

cap = cv2.VideoCapture('cr.mp4') # load the video

my_file = open("data.txt", "r")
data = my_file.read()
class_list = data.split("\n") # read the class labels from the file
print(class_list)

count = 0

while True:    
    ret, frame = cap.read()
    if not ret:
        break  # Exit the loop if the video has ended

    count += 1
    if count % 3 != 0: # ensures that every 3rd frame is processed
        continue

    frame = cv2.resize(frame, (1020, 500)) # set the same image size of all the images
    results = model.predict(frame)
    # print(results) # output the detection details
    a = results[0].boxes.data # bounding box data
    px = pd.DataFrame(a).astype("float") # convert the data to a pd dataframe
    # print(px)
    for index, row in px.iterrows():
        print(row)
        x1 = int(row[0]) # x1
        y1 = int(row[1]) # y1
        x2 = int(row[2]) # x2
        y2 = int(row[3]) # y2
        d = int(row[5]) # class label
        c = class_list[d] # get the class label

        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 1) # green bounding box
        cvzone.putTextRect(frame, f'{c}', (x1, y1), 1, 1) # class label

    cv2.imshow("RGB", frame)
    if cv2.waitKey(1) & 0xFF == 27:  # Exit on 'ESC' key
        break

cap.release()  
cv2.destroyAllWindows()
```
